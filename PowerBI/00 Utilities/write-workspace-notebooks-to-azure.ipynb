```python
import os
from azure.storage.blob import BlobClient
from azure.storage.blob import BlobServiceClient
import sempy_labs.notebook as sempy_notebooks

account = '<account_name_here>'
account_url = f"https://{account}.blob.core.windows.net"
my_container_name = "<container_name_here>"

connect_str = 'CONNECTION STRING GOES HERE'
# Create the BlobServiceClient object
blob_service_client = BlobServiceClient.from_connection_string(connect_str)
```

```python
notebooks = sempy_notebooks.list_notebooks()
display(notebooks)
```

```python
for index, row in notebooks.iterrows():
    my_blob_name = row['Notebook Name'] +'.ipynb'
    nb_def = sempy_notebooks.get_notebook_definition(row['Notebook Name'], format = 'ipynb')    
    # Create a blob client using the local file name as the name for the blob
    blob_client = blob_service_client.get_blob_client(container=my_container_name, blob=my_blob_name)
    print('Uploading notebook: ' + row['Notebook Name'] + ' to container: ' + my_container_name)
    # Upload the created file
    blob_client.upload_blob(nb_def.encode('utf-8'))
```
